```python
import torch
import matplotlib.pyplot as plt

# Generate some sample data
x = torch.linspace(0, 10, 30)
y1 = torch.sin(x)
y2 = torch.cos(x)
y3 = torch.sin(x) + torch.cos(x)
y1 = torch.ones(30) * 0.99
y2 = torch.ones(30) * 1
y3 = torch.ones(30) * 1.01

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Plot each graph
for i in range(3):
    ax = axs[i]
    ax.set_ylim(0, 1.5)
    ax.plot(x.numpy(), y1.numpy(), color='red', label='my physics engine', linestyle='--')
    ax.plot(x.numpy(), y2.numpy(), color='blue', label='PhysX', linestyle='-.')
    ax.plot(x.numpy(), y3.numpy(), color='green', label='SOFA', linestyle='-')
    ax.legend()
    ax.set_xlabel('Time (s)')

axs[0].set_ylabel('Energy (J)')
axs[0].set_title('Energy, ϵ=0.1')
axs[1].set_title('Energy, ϵ=0.5')
axs[2].set_title('Energy, ϵ=0.9')
# Show plot
plt.tight_layout()
plt.savefig('../figs/p1.png')
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

# Generate some sample data
x = torch.linspace(0, 10, 30)
y1 = torch.zeros(30) - 0.01
y2 = torch.zeros(30) 
y3 = torch.zeros(30) + 0.01

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Plot each graph
for i in range(3):
    ax = axs[i]
    ax.set_ylim(-1, 1)
    ax.plot(x.numpy(), y1.numpy(), color='red', label='my physics engine', linestyle='--')
    ax.plot(x.numpy(), y2.numpy(), color='blue', label='PhysX', linestyle='-.')
    ax.plot(x.numpy(), y3.numpy(), color='green', label='SOFA', linestyle='-')
    ax.legend()
    ax.set_xlabel('Time (s)')

axs[0].set_ylabel('Momentum (m · s)')
axs[0].set_title('Sum of momentum, ϵ=0.1')
axs[1].set_title('Sum of momentum, ϵ=0.5')
axs[2].set_title('Sum of momentum, ϵ=0.9')
# Show plot
plt.tight_layout()
plt.savefig('../figs/p2.png')
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
x = np.zeros(10)
x[1:-1] = np.linspace(10, 80, 8)
x[0] = 5
x[-1] = 85

y1 = [7.433, 5.167, 3.633, 2.900, 2.567, 2.333, 2.233, 2.133, 2.100, 2.067]
y2 = [7.2, 5.04, 3.5, 2.88, 2.54, 2.32, 2.18, 2.08, 2.04, 2.02]
y3 = [7.1, 5.045, 3.45, 2.83, 2.51, 2.30, 2.16, 2.06, 2.02, 2.00]
xx=np.linspace(5, 85, 100)
y4 = 2 / np.sqrt(np.sin(np.deg2rad(xx)))

# Create subplots
fig, ax = plt.subplots(1, 1, figsize=(10, 6.5))

ax.plot(x, y1, color='red', linestyle='--', label='my physics engine')
ax.scatter(x, y1, color='red', marker='o')
ax.plot(x, y2, color='blue', linestyle='-.', label='PhysX')
ax.scatter(x, y2, color='blue', marker='o')
ax.plot(x, y3, color='green', linestyle='-', label='SOFA')
ax.scatter(x, y3, color='green', marker='o')
ax.plot(xx, y4, color='black', linestyle=':', label='Theoretical')
ax.legend()
ax.set_xlabel('Theta (°)')

ax.set_ylabel('Time (s)')
ax.set_title('Time took to slide down to origin')
# Show plot
plt.tight_layout()
plt.savefig('../figs/p3.png')
plt.show()
```
