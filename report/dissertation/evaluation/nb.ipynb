```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x_values = [5, 10, 20, 30, 40, 50, 60, 70, 80, 85]
y_values = [7.2, 5.04, 3.5, 2.88, 2.54, 2.32, 2.18, 2.08, 2.04, 2.02]

# Compute function values
x = np.linspace(0, 90, 100)  # Generate x values from 0 to 90
y_function = 2 / np.sqrt(np.sin(x * np.pi / 180))  # Compute y values for the function

# Plot the line for the data points
plt.plot(x_values, y_values, label='Data points')

# Plot the function
plt.plot(x, y_function, label=r'$y = \frac{2}{\sqrt{\sin(x \cdot \pi / 180)}}$')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
