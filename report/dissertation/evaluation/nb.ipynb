```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Generate some sample data
x = [5, 10, 15, 20, 25, 30, 50, 80]
y = [44, 293, 876, 2031, 4000, 7099, 40035.5, 195102]
x = np.array(x) * 1.0
y = np.array(y) * 1.0
y /= x
y /= 30
xx = np.linspace(0, 80, 100)
yy = k_fit * np.power(xx, t_fit)

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.set_ylim(0, 100)
ax.set_xlim(0, 90)
ax.plot(x, y, color='red', linestyle='--',label="Data")
ax.scatter(x, y, color='red', marker='o')
ax.plot(xx, yy, color='black', linestyle='-',label="y = k * x^2")
ax.legend()
ax.set_xlabel('n')
ax.set_ylabel('Time (ms)')
ax.set_title('Average running time')

# Annotate each data point with its y value
for i, (xi, yi) in enumerate(zip(x, y)):
    ax.annotate("{:.2f}".format(yi), (xi, yi), textcoords="offset points", xytext=(0,10), ha='center')

# Show plot
plt.tight_layout()
plt.savefig('../figs/p5.png')
plt.show()

print((33.3 / k_fit) ** (1 / t_fit))
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the function to fit (y = k * x^t)
def func(x, k, t):
    return k * np.power(x, t)

# Sample data
x_data = x
y_data = y

# Perform curve fitting
popt, pcov = curve_fit(func, x_data, y_data)

# Extract the fitted parameters
k_fit, t_fit = popt

# Generate fitted curve
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = func(x_fit, k_fit, t_fit)

# Plot the original data and the fitted curve
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit, color='red', label=f'Fitted curve: y = {k_fit:.2f} * x^{t_fit:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve Fitting')
plt.legend()
plt.grid(True)
plt.show()

print(f'Fitted parameters: k = {k_fit}, t = {t_fit}')
```
